```python
# For matplotlib, the angles have to be in radians, so convert them here
# Could also use np.deg2rad()
testData["plotThetaFromRad"] = testData["plotThetaFrom"] * (np.pi / 180)
testData["plotThetaToRad"] = testData["plotThetaTo"] * (np.pi / 180)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(testData["plotThetaToRad"], testData["plotRadTo"])
ax.set_rmax(7)
ax.set_rticks([1, 2, 3, 4, 5, 6,7])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)
# Needed to avoid a fixed-formatter warning
# Goes from 0 to 2*pi - (1/8*pi) in 1/8pi increments
ax.set_xticks([i*(np.pi/8) for i in range(16)])

#print(ax.get_xticks().tolist())
ax.set_xticklabels(['2012', '2011', '2010', '2009', '2008', '', '', '2021','2020','2019','2018','2017','2016','2015','2014', '2013'])
ax.set_title("A line plot on a polar axis", va='bottom')
#plt.show()
```
